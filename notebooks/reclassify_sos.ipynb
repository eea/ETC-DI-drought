```python
import os
import rasterio
import numpy as np
from rasterio.enums import Resampling
import re

input_path=r'C:\Users\andr3581\OneDrive\Documents\mosaics_mrvvp\dsos\sos_s1'
output_path=r'C:\Users\andr3581\OneDrive\Documents\mosaics_mrvvp\dsos\reclass_sos_s1'
#for reclassifying SOS
#minus then X-n366
#plus then X-n999

for f in os.listdir(input_path):
    if f.endswith('.tif'):
        
        print(f)
        
        #find year based on filename
        reference_year = [int(s) for s in re.findall(r'-?\d+\.?\d*', f)][0]

        
        with rasterio.open(os.path.join(input_path,f)) as src:    
            # Read as numpy array
            array = src.read()
            profile = src.profile

            

            ##2000 is different than all other years
            if reference_year==2000:
                #print('input',np.unique(array))
        
                array=np.where(array<0, array+31706, array)
                #print('output',np.unique(array))
            else:
                #print('input',np.unique(array))
                reference_year_split=int(str(reference_year)[2:4])

                
                before_reference_year=int(str(reference_year_split-1)+str(366))
                after_reference_year=int(str(reference_year_split-1)+str(999))

                if len(str(reference_year_split))==1:
                    inbetween_year_value=int(str(reference_year_split)+('000'))
                elif len(str(reference_year_split))==2:
                    inbetween_year_value=int(str(reference_year_split)+('000'))
                
                #for year 2002
                array=np.where((array<inbetween_year_value) & (array!=0), array-before_reference_year, array)
                array=np.where(array>inbetween_year_value, array-after_reference_year, array)
        
            with rasterio.open(os.path.join(output_path,f), 'w', **profile) as dst:
                # Write to disk

                dst.write(array)
                #make pyramids
                factors = [2, 4, 8, 16, 32]
                dst.build_overviews(factors, Resampling.average)
                dst.update_tags(ns='rio_overview', resampling='average')
                        
        
```
